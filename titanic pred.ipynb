```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
scaling=StandardScaler()
```

```python
x_train_scaled=pd.DataFrame(scaling.fit_transform(x_train[["Age","Fare"]]),columns=["Age","Fare"])
```

```python
x_test_scaled=pd.DataFrame(scaling.transform(x_test[["Age","Fare"]]),columns=["Age","Fare"])
```

```python
x_train.drop(["Age",'Fare'],axis=1,inplace=True)

x_test.drop(["Age",'Fare'],axis=1,inplace=True)
```

```python
x_train.head(5)
```

```python
x_train_scaled=pd.concat([x_train,x_train_scaled],axis=1)
x_test_scaled=pd.concat([x_test,x_test_scaled],axis=1)
```

```python
x_train_scaled.head(4)
```

```python
x_test_scaled.head(4)
```

# Training and Testing

```python
knn_model=knn(n_neighbors=5)
knn_model.fit(x_train_scaled,y_train)
decisiontree_model=DecisionTreeClassifier()
decisiontree_model.fit(x_train_scaled,y_train)
logistic_model=LogisticRegression(max_iter=1000)
logistic_model.fit(x_train_scaled,y_train)
randomforest_model=RandomForestClassifier()
randomforest_model.fit(x_train_scaled,y_train)
```

```python
models=[knn_model,decisiontree_model,logistic_model,randomforest_model]
y_pred=[]
for i in range(4):
    y_predicted=models[i].predict(x_test_scaled)
    y_pred.append(y_predicted)
    
```

```python
acc_score=[]
from sklearn.metrics import accuracy_score
for i in range(4):
#     acc_score.append(accuracy_score(y_test,y_pred[i]))
    print(f"'{models[i]}' -> {accuracy_score(y_test,y_pred[i])}")
```

```python
from sklearn.metrics import confusion_matrix
for i in range(4):
    
    print(models[i])
    print(confusion_matrix(y_test,y_pred[i]))
```
